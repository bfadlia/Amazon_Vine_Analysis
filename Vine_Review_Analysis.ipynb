```python
import pandas as pd
```

```python
vine_df = pd.read_csv('vine_table.csv')
vine_df.head(5)
```

```python
totals_df = vine_df[vine_df['total_votes'] >=20]
totals_df
```

```python
helpful_df= totals_df.loc[(totals_df["helpful_votes"]/totals_df["total_votes"]>= 0.5)]
helpful_df
```

```python
paid_vine_df= helpful_df[helpful_df['vine'] == 'Y']
paid_vine_df
```

```python
unpaid_vine_df= helpful_df[helpful_df['vine'] == 'N']
unpaid_vine_df
```

```python
total_paid = len(paid_vine_df)
total_paid
```

```python
total_unpaid = len(unpaid_vine_df)
total_unpaid
```

```python
count_paid_five = len(paid_vine_df[paid_vine_df['star_rating']==5])
count_paid_five
```

```python
count_unpaid_five = len(unpaid_vine_df[unpaid_vine_df['star_rating']==5])
count_unpaid_five
```
